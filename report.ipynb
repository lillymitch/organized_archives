## Image Preprocessing
Several preprocessing and augmentation techniques were applied to prepare the images for training. All images were transformed using PyTorch and torchvision preprocessing utilities which included image resizing, tensor conversion, and image normalization. The images were resized to a consistent input size so they could be processed uniformly by the neural network architectures. We then normalized the images, helping stabilize training and improve convergence during optimization. Normalization values were selected to match ImageNet preprocessing standards, which was particularly important for the transfer learning model. Random cropping was applied as a form of data augmentation. This technique randomly selects and and crops different portions of an image before resizing the crop to the final input size. It also helps decrease model sensitivity to exact image positioning and encourages it to learn more general visual features rather than memorizing specific image layouts. Random horizontal flipping was used to expose the model to additional visual variations without requiring additional data collection. This helps improve generalization performance and reduces overfitting. Several challenges emerged during data preparation: some artistic styles contained significantly more examples than others, the quality of artwork images varied, many artistic styles shared overlapping visual characteristics, and computational limitations required restricting the total number of processed examples. Despite these limitations, the final dataset provided a strong foundation for training and comparing multiple models for classification.

````python
## Training Transforms
from torchvision import transforms

train_processor = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.RandomResizedCrop(224, scale=(0.7, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])

test_processor = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    )
])
````

## Logistic Regression Baseline Model

Our first model is a Logistic Regression (LR) classifier. This serves as a baseline model that allows us to compare more advanced neural network approaches against a simpler linear method.

Although LR is computationally efficient and easy to interpret, it struggles with image data because it cannot naturally capture spatial relationships or hierarchical visual patterns. The model flattened the image tensor into a one dimensional vector before classification.

````python
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt

# Define a simple Logistic Regression model for image data
class LogisticRegression(nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        # Input image is 3x224x224. Flatten to 3 * 224 * 224 features.
        self.flatten = nn.Flatten()
        self.linear = nn.Linear(3 * 224 * 224, num_classes)

    def forward(self, x):
        x = self.flatten(x)
        return self.linear(x)

# Define the cross-entropy loss function
def cross_entropy_loss(outputs, targets):
    return F.cross_entropy(outputs, targets)
````

## Convolutional Neural Network

Our second model is a custom CNN implemented in PyTorch. CNNs are well-suited for image classification tasks because convolutional layers can automatically learn local visual features such as edges, shapes, textures, and color patterns.

Our CNN architecture includes:
- convolutional layers
- ReLU activation functions
- pooling layers
- fully connected classification layers

The CNN extracts visual features directly from the artwork images and learns hierarchical representations during training.

````python
class CNN(nn.Module):
    """
    Extracts features from input images using convolutional,
    ReLU activation, and pooling layers. The model then reduces
    feature map dimensions with adaptive average pooling before
    using fully connected layers for classification.

    Args:
        num_classes (int): Number of output classes

    Methods:
        forward(x)
    """
    def __init__(self, num_classes):
        super().__init__()

        #the image passes through these layers in order
        #convolution layers learn textures and visual patterns
        #relu adds nonlinearity
        #max pooling reduces image size while keeping important features
        self.pipeline = nn.Sequential(

            nn.Conv2d(3, 16, 5),
            nn.MaxPool2d(2, 2),
            nn.ReLU(),

            nn.Conv2d(16, 32, 3),
            nn.MaxPool2d(2, 2),
            nn.ReLU(),

            nn.Conv2d(32, 64, 3),
            nn.MaxPool2d(2, 2),
            nn.ReLU(),

            #adaptive pooling reduces each feature map to 1x1
            #linear layers use learned image features to predict the class
            nn.AdaptiveAvgPool2d((4, 4)), #changed to 4x4
            nn.Flatten(),
            nn.Linear(64*4*4, 128), #changed to 4x4
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        return self.pipeline(x)
````

### ResNet18 Transfer Learning

Our final model uses transfer learning with ResNet18. Transfer learning allows a model pretrained on a large dataset to be adapted to a new task.

ResNet18 is originally trained on ImageNet, which contains millions of labeled images across many categories. Instead of training an entirely new network from scratch, we reuse pretrained convolutional layers and replace the final classification layer to match our artwork style categories.

This approach provides several advantages such as faster training, better feature extraction, improved performance on limited datasets, and reduced overfitting. Transfer learning is particularly effective because the pretrained model had already learned useful low-level image features such as edges, textures, and shapes.

````python
## ResNet18 transfer learning
from torchvision import models
import torch.nn as nn

def build_transfer_model(num_classes):
  model = models.resnet18(pretrained=True)

  # freeze most layers
  for param in model.parameters():
    param.requires_grad = False

  # unfreeze last layer block (for better performance)
  for param in model.layer4.parameters():
    param.requires_grad = True

  # replace final layer
  model.fc = nn.Linear(model.fc.in_features, num_classes)

  return model
````

## Training Procedure and Optimization

All of our models are trained using Cross Entropy Loss. We experiment with multiple optimizers to compare training performance and convergence behavior. Optimizer comparison is vital because different optimization algorithms can significantly affect convergence speed, training stability, and final accuracy.

Training is performed using PyTorch with GPU acceleration when CUDA is available.


````python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
````

````python
class Trainer:

    def __init__(self, model, train_loader, val_loader, test_loader, optimizer, loss_fn, num_classes, device):
        """
        stores all variables needed for training/testing

        """

        self.model = model.to(device)

        self.train_loader = train_loader
        self.val_loader = val_loader
        self.test_loader = test_loader

        self.optimizer = optimizer
        self.loss_fn = loss_fn

        self.num_classes = num_classes
        self.device = device

        #stores loss/accuracy values over time
        self.history = {
            "train_loss": [],
            "val_loss": [],
            "train_acc": [],
            "val_acc": []
        }


    def evaluate(self, data_loader):
        """
        evaluates the model on a dataset
        and computes accuracy/loss

        params:
            data_loader: dataset loader to evaluate on

        returns:
            accuracy.item(): classification accuracy
            average_loss: average loss across dataset
            conf_mat: confusion matrix
          """
        #puts model into evaluation mode
        self.model.eval()

        #confusion matrix
        conf_mat = torch.zeros(
            (self.num_classes, self.num_classes),
            dtype=torch.int32
        )

        total_loss = 0

        #turns off gradient calculations
        with torch.no_grad():

            for images, labels in data_loader:

                images = images.to(self.device)
                labels = labels.to(self.device)

                outputs = self.model(images)

                #computes loss
                loss = self.loss_fn(outputs, labels)
                total_loss += loss.item() * images.size(0)

                #finds predicted class
                predicted_labels = torch.argmax(outputs, dim=1)

                #fills confusion matrix
                for i in range(len(labels)):

                    true_label = labels[i].item()
                    predicted_label = predicted_labels[i].item()

                    conf_mat[true_label, predicted_label] += 1

        #computes overall accuracy
        accuracy = torch.diag(conf_mat).sum() / conf_mat.sum()
        average_loss = total_loss / len(data_loader.dataset)

        return accuracy.item(), average_loss, conf_mat


    def train(self, num_epochs=1000):
        """
        trains the model for a given
        number of epochs
        """

        print("starting training...")

        for epoch in range(num_epochs):

            #puts model into training mode
            self.model.train()

            epoch_total_loss = 0.0

            for images, labels in self.train_loader:

                images = images.to(self.device)
                labels = labels.to(self.device)

                #clears old gradients
                self.optimizer.zero_grad()

                outputs = self.model(images)

                loss = self.loss_fn(outputs, labels)
                loss.backward()

                #updates weights
                self.optimizer.step()

                epoch_total_loss += loss.item() * images.size(0)

            #evaluates model after each epoch
            training_accuracy, training_loss, _ = self.evaluate(self.train_loader)

            validation_accuracy, validation_loss, _ = self.evaluate(self.val_loader)

            #stores results
            self.history["train_loss"].append(training_loss)
            self.history["train_acc"].append(training_accuracy)

            self.history["val_loss"].append(validation_loss)
            self.history["val_acc"].append(validation_accuracy)

            print("epoch:", epoch + 1)
            print("training accuracy:", round(training_accuracy, 4))
            print("validation accuracy:", round(validation_accuracy, 4))

        print("training finished.")

        return self.history


    def test(self):
        """
        tests the final trained model
        on the test dataset
        """

        print("starting testing...")

        test_accuracy, test_loss, test_confusion_matrix = self.evaluate(
            self.test_loader
        )

        print("test accuracy:", round(test_accuracy, 4))
        print("test loss:", round(test_loss, 4))

        print("test confusion matrix:")
        print(test_confusion_matrix)

        print("testing finished.")

        return test_accuracy, test_loss, test_confusion_matrix
````

## Optimizer Comparison

We compare several optimizers and learning rates to evaluate differences in convergence behavior and training stability. The optimizers we test include standard gradient descent, momentum, and Adam.

These experiments help compare how optimizer selection affects convergence speed and model performance during training.

````python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from typing import Callable, Dict

def test_optimizer(model_builder_function: Callable[[], nn.Module], n_epochs: int = 10) -> tuple[str, Dict[str, Dict[str, any]]]:
    """
    Compares the training loss curves of different optimizers and learning rates
    for a given model architecture.

    Args:
        model_builder_function: A callable that returns a new instance of the model.
                                This ensures each optimizer starts with an identical
                                model state.
        n_epochs: The number of training epochs to run for each optimizer.

    """

    #learning rates for comparison: 'small' (standard) and 'large' (aggressive)
    small_alpha = 1e-4
    large_alpha = 1e-3


    #testing configuration for different optimizers to be tested
    #every entry specifies the optimizer class, learning rate, and momentum (if applicable)
    optimizers_config: Dict[str, Dict[str, any]] = { # 'any' is used for the optimizer_class type
        "Gradient descent": {'lr': small_alpha, 'momentum': 0.0, 'optimizer_class': torch.optim.SGD},
        "Gradient descent (high learning rate)": {'lr': large_alpha, 'momentum': 0.0, 'optimizer_class': torch.optim.SGD},
        "Momentum": {'lr': small_alpha, 'momentum': 0.9, 'optimizer_class': torch.optim.SGD},
        "Momentum (high learning rate)": {'lr': large_alpha, 'momentum': 0.9, 'optimizer_class': torch.optim.SGD},
        "Adam": {'lr': small_alpha, 'optimizer_class': torch.optim.Adam},
        "Adam (high learning rate)": {'lr': large_alpha, 'optimizer_class': torch.optim.Adam},
    }

    losses_dict: Dict[str, list[float]] = {} # Dictionary to store loss history for each optimizer
    final_losses = {}

    #iterates through each optimizer configuration
    for name, config in optimizers_config.items():

        model = model_builder_function()
        optimizer_params = {'lr': config['lr']} #creates optimizer instance

        if 'momentum' in config:
            optimizer_params['momentum'] = config['momentum']

        optimizer = config['optimizer_class'](model.parameters(), **optimizer_params)

        #initializes trainer for this config
        current_trainer = Trainer(
            model=model,
            train_loader=train_loader,
            val_loader=val_loader,
            test_loader=test_loader,
            optimizer=optimizer,
            loss_fn=cross_entropy_loss,
            num_classes=num_classes,
            device=device
        )

        #finds average loss at current epoch using Trainer.evaluate method
        _, initial_avg_epoch_loss, _ = current_trainer.evaluate(train_loader)
        print(f"Initial loss for {name}: {initial_avg_epoch_loss:.4f}")

        losses_dict[name] = []
        losses_dict[name].append(initial_avg_epoch_loss)

        #training loop
        for epoch in range(n_epochs):
            current_trainer.model.train()

            for images, labels in train_loader:
                images, labels = images.to(device), labels.to(device)
                current_trainer.optimizer.zero_grad()

                outputs = current_trainer.model(images)

                loss = cross_entropy_loss(outputs, labels)
                loss.backward()

                current_trainer.optimizer.step()

            #finds average loss at current epoch using Trainer.evaluate method
            _, avg_epoch_loss, _ = current_trainer.evaluate(train_loader)

            losses_dict[name].append(avg_epoch_loss)
            final_losses[name] = avg_epoch_loss


    # --- Plotting Results ---
    fig, axarr = plt.subplots(1, 3, figsize=(15, 5), sharey=True) # Increased figsize for better readability
    fig.suptitle("Training Loss for Different Optimizers", fontsize=16) # Main title for the figure

    optimizer_types = ["Gradient descent", "Momentum", "Adam"]
    ax_map = {opt_type: i for i, opt_type in enumerate(optimizer_types)}

    #ChatGPT helped me write this part but I wrote the starter code myslef

    for name, losses in losses_dict.items():
        # Determine the subplot index based on optimizer type
        current_optimizer_type = name.split(" (")[0]
        ax_idx = ax_map.get(current_optimizer_type)

        if ax_idx is None:
            continue # Skip if optimizer type not recognized

        # Determine label and linestyle based on learning rate
        is_high_lr = "high learning rate" in name
        plot_label = f"{'High LR' if is_high_lr else 'Standard LR'}"
        alpha_val = large_alpha if is_high_lr else small_alpha
        linestyle = '-' if is_high_lr else '--'

        axarr[ax_idx].plot(losses, label=r"$\alpha = $" + f"{alpha_val:.0e} ({plot_label})", linestyle=linestyle)
        axarr[ax_idx].set_title(current_optimizer_type, fontsize=12)
        axarr[ax_idx].grid(True, linestyle='--', alpha=0.7) # Add grid for readability

    # Set common labels and legend for each subplot
    for ax in axarr:
        ax.set_xlabel("Epoch", fontsize=10)
        ax.legend(fontsize=9) # Add legend to each subplot for clarity
        ax.tick_params(axis='both', which='major', labelsize=8)

    axarr[0].set_ylabel("Average Training Loss", fontsize=10) # Only first subplot needs Y-label

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
    plt.show()

    best_optimizer = min(
        final_losses,
        key=final_losses.get
    )

    print("\nBest optimizer:", best_optimizer)

    return best_optimizer, optimizers_config
````

## Training, Evaluation, and Confusion Matrix

````python
import pandas as pd
import torch
import matplotlib.pyplot as plt
from typing import Dict # Add this import for type hinting

def train_and_test_model(model_name, model, optimizer_config: Dict[str, any], num_epochs=5):

    print("\n" + "="*50)
    print(model_name)
    print("="*50)

    model = model.to(device)

    # Extract optimizer class and learning rate from config
    optimizer_class = optimizer_config['optimizer_class']
    lr = optimizer_config['lr']
    optimizer_params = {'lr': lr}

    if 'momentum' in optimizer_config:
        optimizer_params['momentum'] = optimizer_config['momentum']

    optimizer = optimizer_class(
        filter(lambda p: p.requires_grad, model.parameters()),
        **optimizer_params
    )

    trainer = Trainer(
        model=model,
        train_loader=train_loader,
        val_loader=val_loader,
        test_loader=test_loader,
        optimizer=optimizer,
        loss_fn=cross_entropy_loss,
        num_classes=num_classes,
        device=device
    )

    history = trainer.train(num_epochs=num_epochs)

    test_accuracy, test_loss, test_confusion_matrix = trainer.test()

    return {
        "model": model_name,
        "test_accuracy": test_accuracy,
        "test_loss": test_loss,
        "confusion_matrix": test_confusion_matrix,
        "history": history
    }
````

## Error Analysis and Limitations

Several limitations affected overall model performance. Some artistic styles contained significantly more examples than others, leading to class imbalance during training. In addition, many artistic styles shared overlapping visual characteristics, making classification more difficult even for the CNN and ResNet18 models.

The dataset size was also limited because of computational and time constraints. Training on a larger dataset may have improved model generalization and reduced overfitting. Another limitation was image quality inconsistency across the dataset, since some artwork images were lower resolution or contained different lighting and framing conditions.

Although transfer learning improved performance substantially, even the strongest models occasionally confused visually similar styles such as Realism and Baroque.

## Ethical and Social Implications

Machine learning systems for artwork classification could assist museums, researchers, and digital archives in organizing large collections more efficiently. Automated tagging and classification systems may help identify unlabeled works or improve searchability across digital collections.

However, automated artwork classification also introduces limitations and ethical concerns. Artistic interpretation is subjective, and machine learning models may oversimplify historical or cultural context. Biases within the dataset may also affect classification outcomes if certain artistic traditions are overrepresented while others are underrepresented.

These systems should therefore be viewed as assistive tools rather than replacements for art historians, curators, or cultural experts.

## Conclusion

This project explored whether machine learning models could classify artistic styles using only image data. We compared a Logistic Regression baseline model, a custom CNN, and a transfer learning model using ResNet18.

The results showed that model complexity significantly affected classification performance. Logistic Regression struggled because linear models cannot effectively capture spatial image features. The CNN improved performance by learning textures and visual patterns directly from images, while ResNet18 achieved the strongest results through transfer learning and pretrained feature extraction.

Overall, the project demonstrated both the strengths and limitations of machine learning for artwork classification and highlighted the usefulness of transfer learning for image-based tasks with limited datasets.
